Παρακάτω φαίνονται σε λίστα και σχηματικά τα αποτελέσματα των αλγορίθμων και από κάτω φαίνεται ο συγκεντρωτικός πίνακας με τους χρόνους εκτέλεσης, ο οποίος δημιουργείται σε ζωντανό χρόνο μετά την εκτέλεση του προγράμματος.

```python
import matplotlib.pyplot as plt
import time
from utils import *
from convex_hull.ConvexHull import ConvexHull, ConvexHullAlgorithms
from geometrical_search.GeometricalSearch import GeometricalSearch, GeometricalSearchAlgorithms
from geometrical_search.common.Rectangle import Rectangle
```

```python
if __name__ == "__main__":
    for iteration in range(1,5):
        random_points = generate_random_points(num_points=100, min_value=0, max_value=1000)

        algorithms = [
            ConvexHullAlgorithms.INCREMENTAL_ALGORITHM,
            ConvexHullAlgorithms.GIFT_WRAPPING,
            ConvexHullAlgorithms.DIVIDE_AND_CONQUER,
            ConvexHullAlgorithms.QUICKHULL
        ]

        convex_hull_object = ConvexHull()
        times_object = {}

        for algorithm in algorithms:
            start_time = time.time()

            convex_hull_object.set_algorithm(algorithm)
            convex_hull_result = convex_hull_object.compute(random_points, False)

            convex_hull_object_list_result = convex_hull_result.print_as_list()
            convex_hull_object_diagram_result = convex_hull_result.plot_diagram()

            end_time = time.time()
            elapsed_time = end_time - start_time

            # Store the elapsed_time in the times_object dictionary
            times_object[str(algorithm)] = elapsed_time

        # Use the helper function to print the table
        print_algorithm_time_table(times_object)
```

Στο παρακάτω κελί θα εκτελέσουμε ξανά τον αυξητικό αλγόριθμο και θα οπτικοποιηθούν τα βήματα σε μορφή διαγραμμάτων. Στην συγκεκριμένη εκτέλεση του αλγορίθμου εισάγουμε 10 σημεία, καθώς ο αριθμός των βημάτων θα είναι ήδη αυξημένος.

<b>Σχολιασμός Αποτελεσμάτων Αλγορίθμων</b>

Αρχικά, παρουσιάζονται συνοπτικά οι πολυπλοκότητες χρόνου των παραπάνω αλγορίθμων: <br />
Αυξητικός Αλγόριθμος Graham Scan: O(n*logn) <br />
Αλγόριθμος Περιτυλίγματος: O(n^2) <br />
Αλγόριθμος Διαίρε και Βασίλευε: O(n*logn) <br />
Αλγόριθμος Quickhull: O(n*logh) <br />

<b>Αυξητικός Αλγόριθμος Graham Scan:</b> <br />
Ο αλγόριθμος αυτός παρουσιάζει σταθερό χρόνο εκτέλεσης σε όλες τις τρεις δοκιμές, με ελαφρύ περιθώριο διακύμανσης. <br />

<b>Αλγόριθμος Περιτυλίγματος:</b> <br />
Η απόδοση του Gift Wrapping είναι επίσης σταθερή, με μικρές αλλαγές στον χρόνο εκτέλεσης ανάμεσα στις διαφορετικές δοκιμές.<br />

<b>Αλγόριθμος Διαίρε και Βασίλευε:</b><br />
Είναι ενδιαφέρον να βλέπουμε ότι οι αλλαγές στα δεδομένα εισόδου μπορούν να επηρεάσουν την απόδοση του αλγορίθμου.

<b>Αλγόριθμος Quickhull:</b><br />
Η απόδοσή του είναι αντίστοιχη με τους άλλους αλγορίθμους. <br />

<b>Συμπεράσματα:</b>
Ενώ η θεωρητική χρονική πολυπλοκότητα μπορεί να προσφέρει μια γενική ιδέα για την απόδοση των αλγορίθμων, οι πρακτικές εφαρμογές και δοκιμές μπορούν να διαφέρουν ανάλογα με τα δεδομένα εισόδου και την υλοποίηση. Στην περίπτωσή μας, όλοι οι αλγόριθμοι εμφανίζονται να έχουν παρόμοιες αποδόσεις για τα συγκεκριμένα δεδομένα, με τον QuickHull να φαίνεται λίγο πιο γρήγορος.
